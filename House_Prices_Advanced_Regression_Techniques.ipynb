```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_set = pd.read_csv("train.csv")
test_set = pd.read_csv("test.csv")

id = test_set.pop('Id')
```

```python
y_train = train_set['SalePrice']
X_train = train_set.drop(columns=['SalePrice'])
```

```python
X_train.head()
```

```python
X_train.info()
```

From the above description, it seems that there are some columns that contain less than 50% non-null values. These columns do not add much value, so we can drop them.

```python
X_train = X_train.drop("Id", axis = 1) # Don't need Id
test_set = test_set.drop("Id", axis = 1) # Don't need Id
```

```python
non_null_percentage = X_train.notnull().mean() * 100
columns_to_drop = non_null_percentage[non_null_percentage < 50].index
X_train = X_train.drop(columns_to_drop, axis = 1)
test_set = test_set.drop(columns_to_drop, axis = 1)
```

```python
X_train.head()
```

```python
X_train.info()
```

### Handling Numerical Values

```python
df_num = X_train.select_dtypes(include=['int','float'])
```

```python
df_num.isnull().sum()
```

LotFrontage, MasVnrArea and GarageYrBit have missing values

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="median")
imputer.fit(df_num)
```

```python
imputer.statistics_
```

```python
X = imputer.transform(df_num)
```

```python
df_num_tr = pd.DataFrame(X, columns=df_num.columns,
                         index=df_num.index)
```

```python
df_num_tr.info()
```

### Handling Categorical Values

```python
df_cat = X_train.select_dtypes(include=["object"])
```

```python
df_cat.head()
```

```python
df_cat.info()
```

```python
df_cat.isnull().sum()
```

We have missing values in the categorical dataframe.
We can deal with these missing vales using the same logic we used to handle the missing values in the numerical, but here we will impute the values with the most frequent ones.

```python
imputer_cat = SimpleImputer(strategy='most_frequent')
X_cat = imputer_cat.fit_transform(df_cat)
df_tr_cat = pd.DataFrame(X_cat,columns=df_cat.columns)
```

Since we have handled the missing values, we need to encode them becasue the ML model can work only on numerical values.

```python
col = df_tr_cat.columns
```

```python
col
```

Encoding Categorical Data

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder

nominal_columns = [
    'MSZoning', 'Street', 'LotShape', 'LandContour', 'Utilities', 'LotConfig',
    'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'HouseStyle',
    'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'Foundation', 'BsmtExposure',
    'Heating', 'CentralAir', 'Electrical', 'PavedDrive', 'SaleType', 'SaleCondition'
]

ordinal_columns = [
    'ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond', 'BsmtFinType1', 'BsmtFinType2',
    'HeatingQC','KitchenQual', 'Functional', 'FireplaceQu', 'GarageType', 'GarageFinish',
    'GarageQual', 'GarageCond'
]

preprocessor = ColumnTransformer(
    transformers=[
        ('nominal', OneHotEncoder(drop='first'), nominal_columns),
        ('ordinal', OrdinalEncoder(), ordinal_columns)
    ])

df_cat_tr_encoded = preprocessor.fit_transform(df_tr_cat)
```

```python
df_cat_tr_encoded=df_cat_tr_encoded.toarray()
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
df_num_tr_std_scaled = std_scaler.fit_transform(df_num_tr)
```

Final Training Set

```python
train_set_final = np.hstack([df_cat_tr_encoded, df_num_tr_std_scaled])
```

Doing the same steps for test set

```python
# Preprocessing for Test Set (apply the same transformations)

# Handle missing values for numerical columns in the test set
df_test_num = test_set.select_dtypes(include=['int', 'float'])

# Apply the same imputer for numerical columns (using the training set's fitted imputer)
df_test_num_imputed = imputer.transform(df_test_num)
df_test_num_tr = pd.DataFrame(df_test_num_imputed, columns=df_test_num.columns, index=df_test_num.index)

# Handle missing values for categorical columns in the test set
df_test_cat = test_set[nominal_columns + ordinal_columns]
```

```python
df_test_cat = df_test_cat[df_tr_cat.columns]
# Apply the same imputer for categorical columns (using the training set's fitted imputer)
df_test_cat_imputed = imputer_cat.transform(df_test_cat)
df_test_cat_tr = pd.DataFrame(df_test_cat_imputed, columns=df_test_cat.columns)

# Encode categorical features using the preprocessor (fitted to the training data)
df_test_cat_encoded = preprocessor.transform(df_test_cat_tr)

# Apply the same scaler for numerical columns (using the training set's fitted scaler)
df_test_num_scaled = std_scaler.transform(df_test_num_tr)
```

```python
print(f"Shape of df_test_cat_encoded: {df_test_cat_encoded.shape}") # Shape of df_test_cat_encoded: (1459, 156)
print(f"Shape of df_test_num_scaled: {df_test_num_scaled.shape}") # Shape of df_test_num_scaled: (1459, 36)
```

```python
df_test_cat_encoded = df_test_cat_encoded.toarray()
```

Final test set

```python
test_set_final = np.hstack([df_test_cat_encoded, df_test_num_scaled])
```

```python
print(f"Shape of train_set_final: {train_set_final.shape}") # Shape of train_set_final: (1460, 192)
print(f"Shape of test_set_final: {test_set_final.shape}") # Shape of test_set_final: (1459, 192)
```

### Training the model

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()

model.fit(train_set_final, y_train)
```

### Predicting Values

```python
predictions = model.predict(test_set_final)
```

```python
predictions
```

```python
from sklearn.model_selection import cross_val_score

# Perform cross-validation and print accuracy
cv_scores = cross_val_score(model, train_set_final, y_train, cv=5)
print(f"Cross-validation scores: {cv_scores}")
print(f"Mean CV score: {cv_scores.mean()}")
```

These scores are very low
